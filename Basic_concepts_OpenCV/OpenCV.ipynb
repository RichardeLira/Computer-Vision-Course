```python
import cv2 as cv 
from matplotlib import pyplot as plt 
```

Basic functions 

```python
# Original image
img = cv.imread('razer.jpg')
img_BGR = cv.cvtColor(img,cv.COLOR_BGR2RGB)
plt.imshow(img_BGR)
plt.show()
```

```python
# Gray image 
img_gray = cv.cvtColor(img,cv.COLOR_BAYER_BG2GRAY)
plt.imshow(img_gray)
plt.show()
```
