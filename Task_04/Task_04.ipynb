> # Considere a imagem “placas-transito.jpg” aplique a transformada de Hough para linhas e círculos (verifique a documentação do OpenCV) de modo as placas presentes nas imagens.Produza uma imagem diferente para cada um dos três modelos de placa.

```python
# read image 
def read_img(path): 
    img = cv2.imread(path)
    img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    return img
```

```python
def show_img(img):
    plt.figure(figsize = (10,6))
    plt.imshow(img, cmap = 'gray')
    plt.axis('off')
    plt.show()
```

> Read Image 

```python
img = read_img(r'C:\Users\richa\OneDrive\Documentos\Computer Vision\Computer-Vision-Course\Task_04\imagens\placas-transito.jpg')
```

> View image 

```python
show_img(img)
```

> # First type of plate


> circle :

applying the blur


```python
img = cv2.medianBlur(img,5)
show_img(img)
```

```python
circles  = cv2.HoughCircles(img,cv2.HOUGH_GRADIENT,1.0001,20,param1=50,param2=137,minRadius=0,maxRadius=0)
circles = np.uint16(np.around(circles))
```

```python
cimg = cv2.cvtColor(img,cv2.COLOR_GRAY2BGR)
```

```python
for i in circles[0, :]:
    # Outer Circle
    cv2.circle(cimg, (i[0],i[1]),i[2],(0,255,0),2)
    # Center of the Circle 
    cv2.circle(cimg,(i[0], i[1]), 2, (0,255,0),3)
```

```python
show_img(cimg)
```

> # Second type of plate


> diamond:

```python
second_type = read_img(r'C:\Users\richa\OneDrive\Documentos\Computer Vision\Computer-Vision-Course\Task_04\imagens\placas-transito.jpg')
second_type_edges = cv2.Canny(second_type,100,200,apertureSize=3)
```

```python
show_img(second_type_edges)
```

```python
lines = cv2.HoughLinesP(second_type_edges,1,np.pi/180,100,minLineLength=80,maxLineGap=10)
```

```python
for line in lines:
    x1,y1,x2,y2 = line[0]
    cv2.line(second_type,(x1,y1), (x2,y2), (0,255,0), 2)
```

```python
show_img(second_type)
```
